# Importing the necessary libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Reading the Data

```python
with open("fra.txt", 'r') as file:
    text = file.read()
```

# Cleaning the Data

```python
sentences = text.split("\n")
sentences = [sent.split("\t")[:-1] for sent in sentences]
sentences
```

```python
#take only first 10000 samples

sentences = sentences[:10000]
```

```python
englishSentences = []
frenchSentences = []

for sent in sentences:
    if len(sent) == 2:
        englishSentences.append(sent[0])
        frenchSentences.append(sent[1])
        
englishSentences = np.array(englishSentences)
frenchSentences = np.array(frenchSentences)
```

```python
englishSentences.shape, frenchSentences.shape
```

```python
#remove punctuations

import string

englishSentences = [s.translate(s.maketrans("","", string.punctuation)).lower().strip() for s in englishSentences]
frenchSentences = [s.translate(s.maketrans("","", string.punctuation)).lower().strip() for s in frenchSentences]
```

```python
#remove numbers and alphanumerics in the data

englishCleaned = []
frenchCleaned = []
for eng_sent, french_sent in zip(englishSentences, frenchSentences):
    if eng_sent.replace(" ", "").isalpha():
        new_french_sent = ''
        for char in french_sent:
            if char.isalpha() or char == " ":
                new_french_sent += char
        englishCleaned.append(eng_sent)
        frenchCleaned.append(new_french_sent)
    
englishCleaned = np.array(englishCleaned)
frenchCleaned = np.array(frenchCleaned)
```

```python
englishCleaned
```

```python
englishCleaned.shape, frenchCleaned.shape
```

```python
frenchCleaned
```

# Encoding the Data

```python
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
```

```python
def createTokenizer(data):
    tokenizer = Tokenizer(oov_token='<UNK>')
    tokenizer.fit_on_texts(data)
    vocab_size = len(tokenizer.word_index) + 1
    return tokenizer, vocab_size
  
def maxLen(data):
    return max([len(line.split()) for line in data])

def encodeData(tokenizer, data, maxlen):
    encodedData = tokenizer.texts_to_sequences(data)
    encodedData = pad_sequences(encodedData, maxlen = maxlen, padding='post')
    return encodedData



inputTokenizer, input_vocab_size = createTokenizer(englishCleaned)
input_maxlen = maxLen(englishCleaned)
input_sequences = encodeData(inputTokenizer, englishCleaned, input_maxlen)


outputTokenizer, output_vocab_size = createTokenizer(frenchCleaned)
output_maxlen = maxLen(frenchCleaned)
output_sequences = encodeData(outputTokenizer, frenchCleaned, output_maxlen)
```

```python
input_sequences.shape, output_sequences.shape
```

```python
input_vocab_size, output_vocab_size
```

```python
#convert the output into one-hot encoded form

from keras.utils import to_categorical
y = [to_categorical(seq, num_classes = output_vocab_size) for seq in output_sequences]
y = np.array(y)
```

```python
y.shape
```

```python
input_sequences.shape
```

```python
#split into train and test

x_train = input_sequences[:9000]
x_test = input_sequences[9000:]

y_train = y[:9000]
y_test = y[9000:]
```

```python
x_train.shape, x_test.shape
```

# Building the Model

```python
from keras.models import Sequential
from keras.layers import Dense, TimeDistributed, LSTM, Embedding, RepeatVector
```

```python
model = Sequential()

#encoder part
model.add(Embedding(input_vocab_size, 256, input_length=input_maxlen))
# model.add(LSTM(256, return_sequences = True))
model.add(LSTM(256))

model.add(RepeatVector(output_maxlen))

#decoder
# model.add(LSTM(128, return_sequences = True))
model.add(LSTM(256, return_sequences = True))
model.add(TimeDistributed(Dense(output_vocab_size, activation='softmax')))

model.summary()
```

```python
#compiling the model

model.compile(optimizer='adam', loss='categorical_crossentropy')
```

```python
#defining callbacks

from keras.callbacks import ModelCheckpoint, EarlyStopping
# es = EarlyStopping(monitor='val_loss', patience=7, min_delta=0.01)
checkpoint = ModelCheckpoint('nmtModel.h5', monitor='val_loss', save_best_only=True, verbose=1, mode='min')
```

```python
model.fit(x_train, y_train, validation_data=(x_test, y_test), callbacks=[checkpoint], epochs=50, batch_size=64)
```

```python
#plotting the training and validation loss with respect to epochs

modelHistory = model.history
epochs = len(modelHistory.history['val_loss'])
plt.plot(range(1, epochs+1), modelHistory.history['val_loss'], label = 'Validation Loss')
plt.plot(range(1, epochs+1), modelHistory.history['loss'], label = 'Training Loss')
plt.legend()
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.show()
```

# Decoding the Output

```python
def mapIndextoWord(index : int) -> str :
    if index == 0:
        return ""
    for word, i in outputTokenizer.word_index.items():
        if index == i:
            return word

#this function will predict the french translation 
#of an english sentence
def predictSentence(sentence : str) -> str :
    
    encodedData = inputTokenizer.texts_to_sequences([sentence])
    encodedData = pad_sequences(encodedData, maxlen = input_maxlen, padding='post')
    
    output = model.predict(encodedData)[0]
    
    indices = np.argmax(output, axis=1)
    
    translation = ''
    for index in indices:
        if index == 0:
            break
        translation += mapIndextoWord(index) + " "
    
    return translation.strip()
```

```python
#this functions prints the translated sentences
#it takes starting index as the first argument and
#the number of unique sentences we want to translate

def printTranslated(start:int, uniqueSentences:int):

    count = 0
    translationMap = {}
    i = start
    
    while count < uniqueSentences:

        src = englishCleaned[i]
        translation = frenchCleaned[i]
        predicted = predictSentence(englishCleaned[i])

        if src not in translationMap:
            translationMap[src] = [[translation], predicted]
            count += 1
        else:
            translationMap[src][0].append(translation)
            
        i += 1

    for src in translationMap:

        prediction = translationMap[src][1]
        if prediction in translationMap[src][0]:
            translation = prediction
        else:
            translation = translationMap[src][0][-1]

        print("Source ==>", src)
        print("Translation ==>", translation)
        print("Predicted ==>", prediction)
        print()
        
printTranslated(7000, 15)
```

# Evaluate the Model

```python
from nltk.translate.bleu_score import corpus_bleu

def evaluateModel(start : int, end : int, dataset : str):

    src, pred = [], []
    prev_english, prev_french = None, None
    
    for index in range(start, end):

        english_sent = englishCleaned[index]
        french_sent = frenchCleaned[index]
        french_translation = predictSentence(english_sent)
        
        if english_sent != prev_english:
            src.append([french_sent.split()])
            pred.append(french_translation.split())
        else:
            src[-1].append(french_sent.split())
            
        prev_english = english_sent
        prev_french = french_sent
        
    print("BLEU scores for the", dataset, 'are')
    print()
    print('BLEU 1-gram:', "{0:.3f}".format(corpus_bleu(src, pred, weights=(1.0, 0, 0, 0))))
    print('BLEU 2-gram:', "{0:.3f}".format(corpus_bleu(src, pred, weights=(0.5, 0.5, 0, 0))))
    print('BLEU 3-gram:', "{0:.3f}".format(corpus_bleu(src, pred, weights=(0.3, 0.3, 0.3, 0))))
    print('BLEU 4-gram:', "{0:.3f}".format(corpus_bleu(src, pred, weights=(0.25, 0.25, 0.25, 0.25))))
```

```python
evaluateModel(0, 9000, 'TRAINING')
```

```python
evaluateModel(9000, len(englishCleaned), 'TESTING')
```
